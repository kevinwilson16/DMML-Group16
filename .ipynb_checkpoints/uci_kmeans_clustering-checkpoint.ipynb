# UCI Breast Cancer – K-Means Clustering (Unsupervised)

This notebook applies **K-Means clustering** to the preprocessed UCI Breast Cancer Wisconsin (Diagnostic) tabular dataset.

**Goals:**
- Use **only features (X)** to fit an unsupervised model (no labels during training).
- Run **K-Means with k = 2** (intended to roughly correspond to benign vs malignant groups).
- After clustering, compare clusters to the true labels **only for evaluation**.
- Save the trained K-Means model in the `models/` folder for later use.


```python
# 1. Imports

import os
from pathlib import Path

import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, classification_report, silhouette_score

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="muted", font_scale=1.1)

print("Libraries imported successfully!")
```

```python
# 2. Paths and data loading

PROJECT_ROOT = Path.cwd()
DATA_DIR = PROJECT_ROOT / "data" / "processed"
MODELS_DIR = PROJECT_ROOT / "models"
MODELS_DIR.mkdir(parents=True, exist_ok=True)

print("PROJECT_ROOT:", PROJECT_ROOT)
print("DATA_DIR exists?:", DATA_DIR.exists())

# Load preprocessed, scaled features
X_train = pd.read_csv(DATA_DIR / "X_train_scaled.csv")
X_test  = pd.read_csv(DATA_DIR / "X_test_scaled.csv")
y_train = pd.read_csv(DATA_DIR / "y_train.csv").values.ravel()
y_test  = pd.read_csv(DATA_DIR / "y_test.csv").values.ravel()

print("X_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)

feature_names = list(X_train.columns)
```

```python
# 3. Fit K-Means on X (unsupervised)

# We do NOT use y_train here – clustering is unsupervised.

K = 2  # intended to roughly correspond to {benign, malignant}

kmeans = KMeans(
    n_clusters=K,
    random_state=42,
    n_init=10
)

kmeans.fit(X_train)  # only features

train_clusters = kmeans.labels_

test_clusters = kmeans.predict(X_test)

print("K-Means trained.")
print("Cluster counts (train):", np.bincount(train_clusters))
print("Cluster counts (test) :", np.bincount(test_clusters))

sil_train = silhouette_score(X_train, train_clusters)
print(f"Silhouette score (train): {sil_train:.4f}")
```

```python
# 4. Map clusters to labels (for evaluation ONLY)

# Clustering was done without labels; now we check how clusters align with true y.
# We assign each cluster to the majority class in the training set.

cluster_to_label = {}
for c in range(K):
    mask = (train_clusters == c)
    # majority label within this cluster
    majority_label = np.round(y_train[mask].mean()) if mask.sum() > 0 else 0
    cluster_to_label[c] = int(majority_label)

print("Cluster → label mapping (based on train set):", cluster_to_label)

# Apply this mapping to train and test clusters
train_pred_labels = np.vectorize(cluster_to_label.get)(train_clusters)
test_pred_labels  = np.vectorize(cluster_to_label.get)(test_clusters)

print("\n=== Train set evaluation (using mapped cluster labels) ===")
print(classification_report(y_train, train_pred_labels, digits=4))

print("\n=== Test set evaluation (using mapped cluster labels) ===")
print(classification_report(y_test, test_pred_labels, digits=4))

cm = confusion_matrix(y_test, test_pred_labels)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Benign (0)", "Malignant (1)"],
            yticklabels=["Benign (0)", "Malignant (1)"])
plt.title("K-Means (k=2) – Test Confusion Matrix (via mapping)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
# 5. Optional: 2D visualization with PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_all = pd.concat([X_train, X_test], axis=0)
X_all_pca = pca.fit_transform(X_all)
all_clusters = np.concatenate([train_clusters, test_clusters])
all_labels = np.concatenate([y_train, y_test])

plt.figure(figsize=(8,6))
scatter = plt.scatter(
    X_all_pca[:,0], X_all_pca[:,1],
    c=all_clusters, cmap="viridis", alpha=0.7, edgecolor="k", s=40
)
plt.title("K-Means Clusters in PCA 2D Space")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(*scatter.legend_elements(), title="Cluster")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,6))
scatter2 = plt.scatter(
    X_all_pca[:,0], X_all_pca[:,1],
    c=all_labels, cmap="coolwarm", alpha=0.7, edgecolor="k", s=40
)
plt.title("True Labels in PCA 2D Space")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(*scatter2.legend_elements(), title="Label (0=Benign,1=Malignant)")
plt.tight_layout()
plt.show()
```

```python
# 6. Save K-Means model

import joblib

model_path = MODELS_DIR / "kmeans_breast_cancer_model.pkl"
joblib.dump(kmeans, model_path)

print("K-Means model saved to:", model_path)
```
